```python
import pandas as pd
```

```python
path = './taxi_zone_lookup.csv'
df = pd.read_csv(path)
df.head
```

```python
dtype = {
    'LocationID': 'int64',
    'Borough': 'str',
    'Zone': 'str',
    'service_zone': 'str',
}

df = pd.read_csv(path,
                dtype=dtype)
```

```python
from sqlalchemy import create_engine
engine = create_engine('postgresql://root:root@localhost:5432/ny_taxi')
```

```python
print(pd.io.sql.get_schema(df, name='', con=engine))
```

```python
df.head(n=0).to_sql(name='taxi_zone_lookup',con=engine,if_exists='replace')
```

```python
from tqdm.auto import tqdm
df_iter = pd.read_csv(
    path,
    dtype=dtype,
    iterator=True, 
    chunksize = 100000
)
```

```python
for df_chunk in tqdm(df_iter):
    df_chunk.to_sql(name='taxi_zone_lookup', con=engine, if_exists='append')
    
```
